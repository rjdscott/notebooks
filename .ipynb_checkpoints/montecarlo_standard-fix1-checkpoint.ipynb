```python
%%cython

import numpy as np
from timeit import default_timer as timer
from matplotlib import pyplot
import math

cimport cython
cimport numpy as np
cimport matplotlib cimport pyplot

cdef extern from "math.h" nogil:
    double exp(double)
    double sqrt(double)
    double log(double)
    double pow(double, double)
    double fmax(double, double)

    cdef double s, k, r, v, t, p
    cdef int i, j, npath, nstep    
    
cdef step_numpy(double t, double prices, double c0, double c1, double noises):
    return prices * exp(c0 * t + c1 * noises)

cdef mc_numpy(paths, t, i, v):
    c0 = i - 0.5 * pow(v,2)
    c1 = v * sqrt(t)
    
    for j in xrange(1, paths.shape[1]):
        prices = paths[:, j - 1]
        noises = np.random.normal(0., 1., prices.size)
        paths[:, j] = step_numpy(t, prices, c0, c1, noises)
        
    # Stock parameter

    s = 20.83
    k = 21.50
    v = 0.021
    r = 0.20
    t = 5. / 12.
    npath = 3000000
    nstep = 100

    # Plotting

    MAX_PATH_IN_PLOT = 50  
    
# the Driver

cdef driver(pricer, do_plot=False):
    paths = np.zeros((npath, nstep + 1), order='F')
    paths[:, 0] = s
    dt = t / nstep
    
    ts = timer()
    pricer(paths, dt, r, v)
    te = timer()
    elapsed = te - ts
    
    st = paths[:, -1]
    PaidOff = np.maximum(paths[:, -1] - k, 0)
    print 'Result'
    fmt = '%20s: %s'
    print fmt % ('Stock Price', np.mean(st))
    print fmt % ('Standard Error', np.std(st) / sqrt(npath),5)
    print fmt % ('paid off', np.mean(PaidOff))
    optionprice = np.mean(PaidOff) * exp(-r * t)
    print fmt % ('Option Price', optionprice)
    
    print 'Performance'
    NumCompute = npath * nstep
    print fmt % ('Mstep/second', '%.2f' % (NumCompute / elapsed / 1e6))
    print fmt % ('Time elapsed', '%.3fs' % (te - ts))
    
    if do_plot:
        pathct = min(npath, MAX_PATH_IN_PLOT)
        for i in xrange(pathct):
            pyplot.plot(paths[i])
        print 'Plotting %d/%d paths' % (pathct, npath)
        pyplot.show()
    return elapsed        
```

```python
numpy_time = driver(mc_numpy, do_plot=True)
```
